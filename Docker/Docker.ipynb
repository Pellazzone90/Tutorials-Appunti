Cos'è Docker?
-------------------------------------

Docker è una piattaforma per la creazione, l'esecuzione e la condivisione di applicazioni. Serve quindi a garantirne la portabilità gestendo in maniera consistente le dipendenze ed i dati inglobando e perimetrando tutti i requisiti necessari in un ambiente isolato.
Questo ambiente è il *container*.


Virtual Machine VS Container:
-------------------------------------
Possiamo pensare ad un *container* come una versione allegerita di una *Virtual Machine*. 
Quali sono le differenze concrete?
Una virtual machine può funzionare solo grazie a *Software Hypervisors* (tipo VirtualBox o VMware).
Su una stessa macchina possiamo avere più macchine virtuali ed ognuna di esse può eseguire applicazioni diverse fondate su dati e dipendenze diverse. Qaul è il problema? Il problema è che una virtual machine necessita di un sistema operativo completo per cui sono relativamente lente e sono molto *resource intensive* dato che ad esse vengono dedicate porzioni delle risorse fisiche della macchina.

Un container fornisce gli stessi vantaggi di isolamento permettendo di eseguire applicativi diversi ma a differenze delle *VM* sono molto più legger. Perchè? Perchè non necessitano di un intero sistema operativo. Questo perchè tutti i container su una singola macchina condividono lo stesos sistema operativo, quello dell'Host. Di conseguenza non necessitano dell'allocazione di specifiche e "gravose" risorse hardware.


Architettura Docker:
--------------------------------------
Docker utilizza un'architettura Client-Server. Questo significa che Docker ha una componente CLIENT che comunica con la componente SERVER tramite REST API.
La componente SERVER, chiamata DOCKER ENGINE è responsabile della costruzione e dell'esecuzione dei Docker Containers.
Tutti i Container processati da un PC condividono il sistema operativo dell'Host, più precisamente ne condividono il Kernel. Il Kernel, come suggerisc eil nome, è il CORE di un sistema operativo, responsabile dell'organizzazione e gestione delle applicazioni e delle risorse hardware. Ovviamente ogni sistema operativo ha il suo specifico Kernel che determina le sue specifiche API, motivo per cui non possiamo eseguire un'applicazione Windows su un sistema operativo Linux, ad esempio.
Di conseguenza, su una macchina Linux potremo eseguire slo Container Linux. Su una macchina Windows però, a partire da windows 10, è presente nativamente sia un kernel Windows che un kernel Linux. Quindi su Windows possiamo eseguire sia i container Windows che quelli Linux.
Per quanto riguarda i sistemi MAC, questi hanno il loro relativo Kernel che è diverso da quello Linux e da quello Windows e che non ha supporto nativo per applicazioni che hostano container come Docker. Di conseguenza Docker per funzionare su sistemi MAC, utilizzando una versione speciale e molto leggera di virtual machine Linux per poter eseguire container Linux.

Workflow di sviluppo:
-----------------------------------
Come facciamo a dockerizzare un'applicazione? Cioè come rendiamo l'applicazione eseguibile in un container? Semplicemente aggiungendo ad essa un Dockerfile, un file di testo che contiene tutte le istruzioni che Docker utilizza per impacchettare l'applicazione in una *immagine*. L'immagine contiene poi tutte le informazioni necessarie affinchè l'applicazione sia eseguibile nel container.
Una volta che otteniamo l'immagine dell'appliazione, indichiamo a docker di inizializzare un container partendo dall'immagine in questione, è per questo che intendiamo il container come un processo. Resta però un processo particolare poichè ha il suo file system proprietario fornito dall'immagine stessa.
Quando abbiamo l'immagine di un'applicazione in locale, possiamo pusharla nel registry DockerHub, un deposito di immagini Docker a cui chiunque può accedere. In senso inverso possiamo anche pullare una qualsiasi immagine docker da DockerHub. E' questo il modo in cui possiamo condividere e distribuire virtualmente ovunque le immagini che sviluppiamo.


Dockerizzare un'applicazione:
------------------------------------
Come abbiamo detto un'immagine docker contiene tutte le istruzioni/file/configurazioni necessarie per eseguire un'applicazione. Una volta che abbiamo l'immagine possiamo inizializzare u container da essa. \
Un container è il parallelo di una VM nel senso che fornisce un ambiente isolato che può essere avviato, interrotto e riavviato e per questo definiamo il container come un processo.
Ricordiamo però un processo "speciale" in quanto contiene il suo proprietario filesystem.
Quello che però scriviamo nel filesystem di un container resta sono in quello specifico container. Anche se poi runniamo un altro container a partire da una stessa immagine, esso non conterrà i file o le cartelle che abbiamo creato interagendo con il primo container.

La prima cosa da fare per "dockerizzare" un'applicazione è aggiungere un dockerfile ad essa. \
Quali sono le istruzioni che possiamo inserire nel dockerfile per costruire l'immagine?

 - **FROM**: serve per specificare l'immagine base su cui costruire il resto (tipo ubuntu)

 - **WORKDIR**: specifica la working directory, se viene specificato allora poi tutti i comandi verranno 
   eseguiti nella current working directory così specificata

 - **COPY**:per copiare e aggiungere file e directories

 - **ADD**: per aggiungere file e directory ma tramite un URL o un file zippato(che verrà automaticamente 
   unzippato all'esecuzione)

 - **RUN**: per eseguire i comandi relativi al sistema operativo(echo, cat, touch etc) che vengono eseguiti in fase    costruzione dell'immagine essendo una "built-in instruciton"

 - **CMD**: A differenza del comando RUN si tratta di una "run-time instruciton" serve dunque ad eseguire le istruzioni quando inizializziamo il container. Le istruzioni indicate sotto il comando CMD possono essere inseriti in 2 forme.
	La forma *SHELL*: CMD "comando"
	La forma *EXECUTE*: CMD["comando1", "comando2"] come una lista di comandi
 - In genera la forma EXECUTE è sempre consigliata in quanto la forma SHELL richiede l'avviamento di una shell ulteriore che dipende poi dal sistema operativo dell'immagine.

 - **ENTRYPOINT**: per eseguire i comandi, come per CMD anche ENTRYPOINT ha le sue 2 forme SHELL ed EXECUTE. A cosa serve ENTRYPOINT? In generale quando runniamo un container si possono sempre modificare i comandi di default in fase di avvio, ad esempio eseguendo" docker run nome_container echo ciao" invece di "docker run nome_container". ENTRYPOINT ci garantisce una maggiore robustezza in quanto per sovrascrivere i comandi default di quest'ultimo bisognerebbe aggiungere anche il parametro --entry quando si esegue il "docker run nome_container"

 - **ENV**: per settare le variabili d'ambiente ad esempio l'URL di una API con cui il container deve comunicare esternamente

 - **EXPOSE**: per indicare la porta che deve aprire il container quando in esso è avviata un'applicazione hostata localmente (localmente dal container) come mappare una porta dall'host (il pc) ad una porta nel container?
   Più che un comando è un'indicazione da documentazione poichè il comando non pubblica automaticamente la porta da    cui ascolta il container ma indica che quell'applicazione resterà in ascolto eventualmente sulla porta indicata

 - **USER**: Per indicare i gruppi ed i permessi utenti. Di default Docker esegue le applicazioni con l'utente Root.

Snellire un'immagine Docker:
------------------------------------
Quando andiamo a creare il Dockerfile useremo sicuramente il comando COPY o ADD per aggiungere le directory dell'applicazione che vogliamo dockerizzare. \
Partendo da una specifica immagine root, quella ottenuta dal comando FROM, costruiremo un'immagine che è sicuramente abbastanza pensate sia in termini di MB che in termini di risorse necessarie per buildarla.
Nel buildare la nostra immagine ci andiamo a collegare ad un Docker Engine remoto per cui tutti i dati passano in rete. Se abbiamo un'immagine molto pesante rallenteremo questo processo di creazione. \
Cosa possiamo fare quindi? Escludere qualche file/directory? Sicuramente. Ad esempio, è inutile che andiamo ad includere nell'immagine la directory contenente tutte le dipendenze e le librerie utilizzate per eseguire l'applicazione, basta invece aggiungere un file di requisiti che dica all'immagine nel momento dell'avvio quali librerie e pacchetti installare. \
Di conseguenza vogliamo escludere alcune cartelle da copiare, come fare? Al pari del **.gitignore"** esiste il **.dockerignore**. \
All'interno del file *dockerignore* possiamo elencare i file e le directory che vogliamo escludere dall'elaborazione del docker engine in fase di costruzione dell'immagine, velocizzando l'operazione di "build image" ed ottenendo così un'immagine molto più leggera.

Come ottimizzare la creazione di  immagini:
------------------------------------
Una Docker image è essenzialmente una collezione di *Layer*, cioè piccoli file system che includono solo file modificati/aggiornati. Ogni volta che Docker costruisce un'immagine aggiunge un layer per ogni istruzione del docker file. Docker ha un sistema di ottimizzazione per la costruzione delle immagini, ogni qualvolta deve processare un nuovo layer a seguito di un'istruzione verifica se questa è cambiata o è stata modificata. Se l'istruzione non è cambiata non andra a ricostruire il layer relativo ma utilizzerà quello memorizzato nella sua cache. C'è un problema però, se un layer viene ricostruito a seguito di una modifica, verranno ricostruiti anche tutit i layer successivi a cascata. \
Un approccio intelligente per velocizzare il processo di creazione dell'immagine è quello di separare comandi di installazione di dipendenze ed i comandi per copiare file all'internno dell'immagine. Conviene sempre prima copiare ed installare tutte le dipendeze necessarie e solo dopo aggiungere o copiare i dati relativi all'applicazione. In questo modo Docker riutilizzerà sempre le istruzioni della sua cache nella fase di "settaggio" dell'ambiente dell'applicazione.

Come rimuovere immagini o containers:
------------------------------------
A volte nell'elenco di immagini listato tramite il comando "docker images" possiamo ritrovare delle immagini che non hanno nè nome nè tag. Queste immagini vengono definite *Dangling Images* sono essenzialmente layer che non hanno relazione con un'immagine taggata generati ogni volta che modifichiamo il dockerfile e ribuildiamo un'immagine.
Per rimuoverle dobbiamo utilizzare il comando:\
```console
simonet@pop-os:~$ docker image prune
```
Avremo anche un numero identico di container stoppati relativi alle *Dangling Images*, allo stesso modo rimuoveremo i container con il comando: \
```console
simonet@pop-os:~$ docker container prune
```
(Rimuoverà tutti i container stoppati). \
Per rimuovere una specifica immagine invece dobbiamo eseguire il comando:\
```console
simonet@pop-os:~$ docker rm image ****
```
Riferendoci all'immagine inserendo al posto dei **** il nome o l'ID


Come taggare un'immagine:
------------------------------------
Ogni volta che buildiamo un'immagine docker la tagga con il valore "latest" ma questo è solo una label, non significa affatto che sia "la più recente".\
Per aggiunger eun TAG all'immagine basta aggiungere affianco la nome dell'immagine in fase di creazione:
```console
simonet@pop-os:~$ docker build -t nome_image:tag_a_piacere
```
Si può taggare un'immagine anche dopo averla creata con il comando:
```console
simonet@pop-os:~$ docker image tag nome_image:tag_vecchio nome_image:tag_nuovo
```
**NB:** Una stessa immagine può avere diversi TAG

Come condividere un'immagini su DockerHub:
------------------------------------
Dopo aver creato una repository su DockerHub per pushare un'immagine locale in remoto eseguiamo:
```console
simonet@pop-os:~$ docker login
simonet@pop-os:~$ docker push nome_immage/tag_immagine
```
In questo modo tutti i layer verranno pushati, il primo push sarà lento ma i successvi no poichè sfrutteranno lo stesso principio di cache applicato in fase di creazione.\

Possiamo "condividere" un'immagine anche senza passare da DockerHub, basta salvare l'immagine in un file compresso e "spostarla".
Per farlo bisogna utilizzare il comando:
```console
simonet@pop-os:~$ docker image save -o nome_file.tar(.zip)
simonet@pop-os:~$ docker image load -i nome_file
```


Alcuni comandi Docker utilizzati:
------------------------------------
```console
simonet@pop-os:~$ docker build -t nome_image #per buildare un immagine con nome test_image
simonet@pop-os:~$ docker run nome_image #per runnare l'immagine
simonet@pop-os:~$ docker ps #per visualizzare tutto i container attivi
simonet@pop-os:~$ docker ps -a #per visualizzare tutti i container anche quelli non in esecuzione
simonet@pop-os:~$ docker image -l/ docker images #per listare tutte le immagini presenti in locale
simonet@pop-os:~$ docker stop nome_o_tag_container #per interrompere un container attivo
simonet@pop-os:~$ docker start nome_o_tag_container #per ri-avviar eun container stoppato
simonet@pop-os:~$ docker exec -it nome_image bash #per avviare l'immagine ed interagirvi tramite shell Bash
```
